Flow Control </br>
(1) for loop </br>
Example: list every nucleotides of a sequence

```python
seq_a="ATGCTGCAT"
for i in range(len(seq_a)):
    print(seq_a[i])
```

Flow control <br>
(2) while loop and (3) if-else<br>
Example: print the nucleotide of a sequence until "CA" is encountered encountered

```python
seq_a="ATGCTGCAT"
index=0
while True:
    if seq_a[index:index+2]=="CA":
        break
    else:
        print(seq_a[index])
        index+=1
```

function  <br>
<tab>Think of it as a verb <br>
<tab>A function performes certain tasks <br>

Example: when printing out the elements of a seq, <br>
       also print out its index(position in the seq)

```python
seq1="ACTT"
#define a function
def print_with_index(seq):
    for i in range(len(seq)):
        print("%s:%s" % (i,seq[i]))
#call a function
print_with_index(seq1)
```
